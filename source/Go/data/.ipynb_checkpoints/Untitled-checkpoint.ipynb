```python
import pandas as pd
import numpy as np

def mag(x):
    return math.sqrt(sum(i**2 for i in x))
```

```python
score = pd.read_csv('score_board.csv',index_col = 0)
score
```

```python
score['visit'] = 1
score
```

```python
p_m = pd.read_csv('normalized_item_matrix.csv',index_col = 0)
u_m = pd.read_csv('normalized_user_matrix.csv',index_col = 0)
```

```python
p_m
```

```python
u_m
```

```python
import math
from tqdm import tqdm
```

```python
cos_sim = []
for u, p in tqdm(zip(list(score['u_id']),list(score['p_id']))):
    try:
        if u == 'dhyeok1996' or u=='u1s2&dhyeok1996':
            break
        u_vec = u_m.loc[int(u),:]
    except:
        u_vec = u_m.loc[u,:]
    p_vec = p_m.loc[p,:]
    u_size = mag(u_vec)
    p_size = mag(p_vec)
    if u_size==0 or p_size==0:
        cos_sim.append(0)
    else:
        cos_sim.append(np.dot(p_vec, u_vec)/(p_size*u_size))
```

```python
score = score[score['rating'].notna()]
```

```python
score['cbf_score'] = cos_sim
```

```python
score.sort_values('cbf_score',ascending=False)[:50]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
RMSE = mean_squared_error(score['visit'],score['cbf_score'])**5
RMSE
```

```python
r2_score(score['visit'],score['cbf_score'])
```
