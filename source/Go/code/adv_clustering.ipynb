```python
import pandas as pd
import numpy as np
import modules #직접 만든 모듈파일
import nltk
from gensim.models import KeyedVectors
from sklearn.cluster import KMeans
```

```python
user_df = pd.read_csv('../data/all_users_reviews.csv')
places_df = pd.read_csv('../data/all_places_reviews.csv')
del places_df['review_kr']
places_df.rename(columns={'review_en': 'review'}, inplace=True)
user_df #20706
user_df.dropna(inplace=True) #----20706----
places_df.dropna(inplace=True) #2433
places_df #----1409----
places_df.reset_index(inplace=True) #----1409----
```

```python
review = ''
for i in range(0,len(places_df)):
    review = review + places_df['review'][i]
```

```python
review = review.lower()
```

```python
review = modules.delete_punctuations(review)
```

```python
pos_token = nltk.pos_tag(review.split())
pos_token
```

```python
stop_words = []
```

```python
w2v = KeyedVectors.load_word2vec_format("../models/GoogleNews-vectors-negative300.bin", binary=True)
```

```python
new_df = pd.DataFrame()
for i in pos_token:
    if (i[1] == 'JJ') & (i[0] not in stop_words):
        word=modules.delete_punctuations(i[0]);
        try:
            f_vec = w2v.get_vector(word)
            new_df[word] = f_vec
        except:
            pass
#             print(word, ' not in vocab')
```

# From Here

```python
new_df.to_csv('./pre_w2v', index=False)
```

```python
new_df.shape
```

```python
x = new_df.T
x.shape
```

```python
cluster_count=95
model = KMeans(n_clusters=cluster_count)
```

```python
model.fit(x)
model.predict(x)
Y=x.copy()
Y['kmeans_id'] = model.predict(x)
```

```python
km = list(Y['kmeans_id'])
max_count = 0
max_count_id = ''
for i in range (0, cluster_count):
    if km.count(i) > max_count:
        max_count = km.count(i);
        max_count_id = i;
    print(i,': ', km.count(i))
max_count_id
```

```python
km_df = pd.DataFrame()
km_df['group_'+str(max_count_id)] = pd.Series(list(Y[Y['kmeans_id'] == max_count_id].index))
for i in range(0, cluster_count):
    col_name = str(i)
    words = list(Y[Y['kmeans_id'] == i].index)
    km_df[col_name] = pd.Series(words)
```
