```python
from selenium import webdriver as wd
from selenium.webdriver import ActionChains as ac
from time import sleep
import pandas as pd
import re
from bs4 import BeautifulSoup as bs
import math
from selenium.webdriver.common.by import By
```

```python
spot_ls = list(pd.read_csv('../data/korea_travel_places.csv')['name'])
```

```python
len(spot_ls)
```

```python
for name in spot_ls:
    if ("(" in name):
        spot_ls[spot_ls.index(name)] = name.split('(')[0]
```

```python
spot_ls=spot_ls[2:]
```

```python
spot_ls
```

```python
#리뷰박스
review_box = "div.location-review-card-Card__ui_card--2Mri0"
user_id = "a.ui_header_link.social-member-event-MemberEventOnObjectBlock__member--35-jC"
review_title = "a.location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT"
rating = "span.ui_bubble_rating"
review_content = "q.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC"
```

```python
def get_reviews(driver, p_id):
    df = pd.DataFrame(columns={'p_id','u_id','u_rate','title','review'})
    try:
        #focusing할 태그 선택
        content=driver.find_element_by_css_selector('div.main_content')
        boxes = content.find_elements_by_css_selector(review_box)

        for box in boxes:
            userId = box.find_element_by_css_selector(user_id).text
            title = box.find_element_by_css_selector(review_title).text
            rate = get_rate(box)
            content = box.find_element_by_css_selector(review_content).text
            df= df.append({'p_id':p_id, 'u_id':userId, 'u_rate': rate, 'title': title, 'review': content}, ignore_index=True)
        return df
    except:
        return;
```

```python
def get_rate(box):
    try:
        box = box.find_element_by_css_selector('div.location-review-review-list-parts-RatingLine__bubbles--GcJvM')
        small_box = box.find_element_by_css_selector('span')
        rate = int(small_box.get_attribute("class")[-2:-1])
        return rate;
    except:
        return '*';
```

```python
#pagination
def click_next_page(driver):
    try:
        driver.find_element_by_css_selector("a.ui_button.nav.next.primary").click()
    except:
        return;
```

```python
def click_to_switch_tap(driver):
    #리뷰로 이동
    driver.find_element_by_css_selector("div.result-title").click()
    #탭 전환
    driver.switch_to_window(driver.window_handles[1])
```

```python
def get_last_review_page(driver):
    try:
        page_num_box=driver.find_element_by_css_selector("div.pageNumbers")
        page_nums=page_num_box.find_elements_by_css_selector("a.pageNum")
        return int(page_nums[len(page_nums)-1].text)
    except:
        return 0;
```

```python
def main():
    #필요한 변수 선언
    index=1503
    for spot_name in spot_ls:
        df = pd.DataFrame(columns={'user_id','rating','title','content'})
        url = f"https://www.tripadvisor.co.kr/Search?q={spot_name}"
        driver = wd.Chrome()
        driver.get(url)
        #가장 첫번째 검색결과
        sleep(3)
        try:
            res_title = driver.find_element_by_css_selector("div.result-title").text       
            print(res_title)
            #todo 결과와 input값 비교 (공백 제거 후)
            if ((spot_name.replace(' ','') in res_title.replace(' ','')) | (res_title.replace(' ','') in spot_name.replace(' ','')) ):
                print('here')
                click_to_switch_tap(driver)
                sleep(1)
                last_index = get_last_review_page(driver)
                print('last_index', last_index)
                sleep(2)
                i=1
                if (last_index != 0):
                    for i in range(1, last_index+1):
                        df_new = get_reviews(driver)
                        df = pd.concat([df, df_new], ignore_index=True)
                        click_next_page(driver)
                        print('now page', i)
                        sleep(2)
                    df['place_name'] = spot_name;
                    df['res_title'] = res_title;
                    df.to_csv(f'./reviews/{index, spot_name}.csv', index = False, header=True)
        except:
            pass
        driver.quit()
        print(df)
        df=[];
        df_new=[];
        index = index+1
        sleep(1)
    return;
```

```python
main()
```
