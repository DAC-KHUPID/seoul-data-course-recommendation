```python
df = df.drop_duplicates(['name'])
```

```python
from pypapago import Translator
ts = Translator()
```

```python
for i in range (len(df)):
        if(str(type(df['reviews'][i])) == "<class 'str'>"):
            file_name = df['reviews'][i]
            review_df = pd.read_csv(f'../koreatour_reviews/{file_name}')
            all_review_kr = ''
            all_review_en = ''
            for j in range (len(review_df)):
                print('init', i, j);
                a_review = review_df['review'][j];
                try:
                    review_en = ts.translate(a_review, source='ko', target='en', verbose=False)
                    print('try_suc', i, j);
                except:
                    try:
                        print('except1', file_name, review_df['Unnamed: 0'][j])
                    except:
                        print('except2', file_name)
                        pass
                all_review_kr = ". ".join((all_review_kr, str(a_review)))
                all_review_en = ". ".join((all_review_en, str(review_en)))
            df['review_kr'][i] = all_review_kr;
            df['review_en'][i] = all_review_en;
            
                        
```
