```python
for place in null_kakao_places:
    df = pd.read_csv(f'../../data/kakao_reviews/{place}')
    all_review = ''
    all_en_review = ''
    p_id = df['p_id'][0]
    for i in range (0, len(df)):
        review = df['review'][i]
        try:
            review_en = ts.translate(review, source='ko', target='en', verbose=False)
        except:
            print(place, i)
            review_en = ''
        all_review = all_review+'. '+review;
        all_en_review = all_en_review+'. '+review_en
    each_place_df = each_place_df.append({'p_id':p_id, 'review':all_review, 'review_en': all_en_review}, ignore_index=True)
```
