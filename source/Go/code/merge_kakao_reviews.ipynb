```python
from os import listdir

filepath ="/Users/ryeonju/code/DAC/kakao_reviews";
```

```python
listdir(filepath)
```

```python
import pandas as pd

review_df=pd.DataFrame(columns={'p_id','u_id','u_rate','review'})
review_df = review_df[['p_id','u_id','u_rate','review']]
for x in listdir(filepath):
    df = pd.read_csv(f'./kakao_reviews/{x}')
    review_df=pd.concat([review_df, df])
```

```python
review_df.reset_index(drop=True, inplace=True)
review_df
```

```python
print('결측치 제거 전 데이터 개수', len(review_df))
review_df['review'].dropna()
print('결측치 제거 후 데이터 개수', len(review_df))
```

```python
review_df.to_csv('kakao_reviews.csv', index =False, header=True)
```

```python
import pandas as pd

kakao_place_df=pd.DataFrame(columns={'p_id','p_name','address','tags', 'open_time', 'open_day', 'category', 'price', 'p_rate', 'link'})
kakao_place_df=kakao_place_df[['p_id','p_name','address','tags', 'open_time', 'open_day', 'category', 'price', 'p_rate', 'link']]
place_ids=['a','b','c','d','e','f','g']

for x in place_ids:
    df = pd.read_csv(f'kakao_places_{x}.csv')
    kakao_place_df=pd.concat([kakao_place_df, df])
    
kakao_place_df.reset_index(drop=True, inplace=True)
kakao_place_df
```

```python
kakao_place_df.to_csv('kakao_places.csv', index = False, header=True)
```
