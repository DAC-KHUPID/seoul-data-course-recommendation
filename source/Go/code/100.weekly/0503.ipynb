```python
import pandas as pd
```

```python
df = pd.read_csv('./togo_new2.csv')
df
```

```python
for i in range (0, len(df)):
    if(pd.isnull(df['reviews'][i]) == False):
        file_path = './koreatour_reviews/'+str(df['reviews'][i])
        file = pd.read_csv(file_path)
        for j in file:
            if j == 'user_id':
                file.rename(columns = {'user_id' : 'u_id', 'content':'review', 'rating':'u_rate'}, inplace = True)
                file['p_id'] = df['id'][i]
                file.to_csv(file_path)
                break;
```
