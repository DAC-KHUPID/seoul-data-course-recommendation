# KAKAO

```python
user_df.sort_values(by=['u_id'], axis=0, inplace=True)
user_df.reset_index(drop=True, inplace=True)
```

```python
user_df.dropna(subset=['u_id'], inplace=True)
```

```python
user_df.drop_duplicates(['u_id', 'p_id'], keep='first', inplace=True)
```

```python
user_df['review_en'] = 0
user_df.reset_index(drop=True, inplace=True)
```

```python
user_df
```

```python
from pypapago import Translator
ts = Translator()
for i in range(0, len(user_df)):
    ko = user_df['review'][i]
    try:
        en = ts.translate(ko, source='ko', target='en', verbose=False)
        user_df['review_en'][i] = en;
        print('printed', i,)
    except:
        print(ko)
        pass
```

```python
for i in range(0, len(user_df)):
    if(user_df['review_en'][i] == 0) :
        print(i)
```

```python
user_df.to_csv('../../data/kakao_reviews_each_user.csv', index=False)
```
