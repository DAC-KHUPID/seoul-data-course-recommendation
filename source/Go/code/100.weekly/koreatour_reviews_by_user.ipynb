```python
import os
import pandas as pd
```

```python
df = pd.DataFrame(columns=['u_id', 'p_id', 'u_rate', 'review', 'review_en'])
df
```

```python
for i in os.listdir("../../data/koreatour_reviews"):
    get_df=pd.read_csv(f"../../data/koreatour_reviews/{i}")
    df = pd.concat([df, get_df])
```

```python
df
```

```python
df.drop_duplicates(['u_id', 'p_id'], inplace=True)
df.sort_values('u_id', inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
merged_df = pd.DataFrame(columns=['u_id','review'])
```

```python
i = 0
while i < len(df):
    print(i)
    u_id = df['u_id'][i]
    review = df['review_en'][i]
    for j in range (i+1, len(df)):
        if (u_id == df['u_id'][j]):
            review = review+'. '+df['review_en'][j]
        else:
            break;
#     print({'u_id': [u_id], 'review': [review]})
    new_data = pd.Series({'u_id': u_id, 'review': review})
    merged_df = merged_df.append(new_data, ignore_index=True)
    i=j
    if(i == 57312):
        break;
```

```python
merged_df.to_csv(f"../../data/koreatour_each_user", index=False )
```
