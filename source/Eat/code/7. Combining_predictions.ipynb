```python
import pandas as pd
import numpy as np
```

```python
user_id = int(input('user id : '))
pd.set_option('display.float_format', lambda x: '%.6f' % x)
```

```python
eat_info =  pd.read_csv('../data/translated_eat_info.csv')
cbf = pd.read_csv('../data/predictions/'+str(user_id)+'_cbf.csv')
ncf = pd.read_csv('../data/predictions/ncf_scores.csv')

eat_info = pd.DataFrame(eat_info, columns=['p_id','name','address'])
eat_info.set_index('p_id',drop=True,inplace=True)
cbf.set_index('p_id',drop=True,inplace=True)
ncf = ncf.rename(columns = {'0':'p_id','1':'ncf_scores'})
ncf['p_id'] = ncf['p_id'].astype(int)
ncf = ncf.set_index('p_id', drop=True).drop('u_id', axis=1)\
```

```python
result = eat_info.join(cbf, how='left').join(ncf, how='left')
result.head()
```
