```python
#eat/data/score_board
eat_ncf = pd.read_csv('../../eat/data/score_board.csv')
for sv_index in range (0, len(sv)):
    u_id = sv['u_id'][sv_index]
    eat_list = sv['Eat'][sv_index]
    try: 
        for eat_item in eat_list:
            if (eat_item == '광장시장 육회자매집'):
                eat_item = '육회자매집'
            p_id = eat[eat_item]
            eat_ncf = eat_ncf.append({'u_id':u_id,'p_id':p_id}, ignore_index=True)
    except:
        print(u_id, eat_list)
        pass
eat_ncf.to_csv('../../eat/data/score_board.csv', index=False)
```

```python
#go/data/ncf_data.csv
go_ncf = pd.read_csv('../../go/data/ncf_data.csv')
for sv_index in range (0, len(sv)):
    u_id = sv['u_id'][sv_index]
    go_list = sv['Go'][sv_index]
    for go_item in go_list:
        if (go_item == '남대문시장(회현시장)'):
            go_item = '남대문시장'
        p_id = go[go_item]
        go_ncf = go_ncf.append({'u_id':u_id,'p_id':p_id}, ignore_index=True)
go_ncf.to_csv('../../go/data/ncf_data.csv', index=False)
```
