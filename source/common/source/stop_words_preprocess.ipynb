```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wd
```

```python
url = "https://www.ranks.nl/stopwords/korean"
```

```python
driver = wd.Chrome()
driver.get(url)
```

```python
tbody = driver.find_element_by_css_selector('tbody').text
stop_words = tbody.split('\n')
```

```python
stop_words
```

```python
file = open('./stop_words2.txt','r')
words = file.read()
file.close()
```

```python
stop_words2 = words.split('\n')[:-1]
stop_words2
```

```python
for i in range (0, len(stop_words2)):
    stop_words2[i] = stop_words2[i].split('\t')[0]
```

```python
stop_words2
```

```python
print(len(stop_words))
print(len(stop_words2))
stop_words = stop_words+stop_words2
stop_words = list(set(stop_words))
```

```python
df = pd.DataFrame({'stop_word': stop_words})
df.to_csv('all_stop_words.csv', index = False, header=True)
```
