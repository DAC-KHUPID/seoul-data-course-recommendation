```python
import pandas as pd
import pprint
pp = pprint.PrettyPrinter(indent=4)
```

```python
file=pd.read_csv("movie_watcha_review.csv",index_col=0)
file.head()
```

```python
user_list=list(set(list(file['user'])))
```

```python
user=pd.DataFrame(data=list(range(6156)), index=range(6156), columns=['user_id'])
user['user']=pd.Series(user_list, index=user.index)
user_id_dict={}
for i in range(len(user)):
    user_id_dict[user['user'][i]]=user['user_id'][i]
user.head()
```

```python
data={
    'movie_id':file['movie_id'],
    'rate':file['rate'],
    'en_review':file['en_review']
}
user_matrix=pd.DataFrame(data, columns=['movie_id','user_id','rate','en_review'], index=range(len(file['movie_id'])))
for i in range(len(user_matrix)):
    user_matrix['user_id'][i]=user_id_dict[file['user'][i]]
user_matrix
```

```python
user_matrix.to_csv('user_matrix.csv')
```
